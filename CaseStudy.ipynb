# Bar Graph

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

df = pd.read_csv (r'C:/Users/user/Desktop/CaseStudy/cars_data.csv')
plt.figure(figsize=(10,9))
plt.style.use('ggplot')
sns.barplot(data=df, x="Buyer Gender", y="Sale Price")
plt.title('bar graph of male vs female buyers participated in the sales.')
plt.xlabel('Buyer Gender')
plt.ylabel('Sales Price')
plt.show()


```

#  Top 5 cars based on their sales price.

```python
import pandas as pd

df = pd.read_csv (r'C:/Users/user/Desktop/CaseStudy/cars_data.csv')
a=df.nlargest(5, ['Sale Price'])
#print (a)
print("top 5 cars based on their sales price")
a[['Make', 'Sale Price']]
```

#  The least 5 cars based on their Resell price
